```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from random import random

# 
fig = plt.figure(figsize=(6, 10), facecolor='w')

#  3D 
ax = fig.add_subplot(projection='3d')

# 
img = cv2.imread('/root/autodl-tmp/lighten-diffusion/23.png')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
x, y, z = img2[:, :, 0], img2[:, :, 1], img2[:, :, 2]

#  3D 
points = ax.scatter(xs=x,  # x 
                    ys=y,  # y 
                    zs=z,  # z 
                    zdir='z',
                    s=1,  # size
                    alpha=0.05,
                    cmap='magma',
                    c=z)

mappable = plt.cm.ScalarMappable(cmap='magma')  # 
mappable.set_array(z)  # z orientation

# 
c1 = ax.set(xlabel='H',
            ylabel='S',
            zlabel='V',  # 
            xticks=np.arange(0, 255, 40),
            yticks=np.arange(0, 255, 40),
            zticks=np.arange(0, 255, 40)
            )

ax.set_xlabel('H', fontsize=15, labelpad=5)
ax.set_ylabel('S', fontsize=15, labelpad=5)
ax.set_zlabel('V', fontsize=11, labelpad=1)  # 
labels = ax.get_xticklabels() + ax.get_yticklabels() + ax.get_zticklabels()

[label.set_fontsize(12) for label in labels]
[label.set_fontweight('bold') for label in labels]

# 
ax.view_init(elev=17, azim=45)

# 
plt.colorbar(mappable, fraction=0.02, cax=ax.inset_axes([1, 0.1, 0.05, 0.8]))  # 
plt.show()
```

```python
    import cv2

    from matplotlib import pyplot as plt
    img = cv2.imread('/root/autodl-tmp/lighten-diffusion/23.png')
    
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = img2[:,:,0], img2[:,:,1], img2[:,:,2]
    hist_h = cv2.calcHist([h],[0],None,[256],[0,256])
    hist_s = cv2.calcHist([s],[0],None,[256],[0,256])
    hist_v = cv2.calcHist([v],[0],None,[256],[0,256])
    plt.plot(hist_h, color='r', label="h")
    plt.plot(hist_s, color='g', label="s")
    plt.plot(hist_v, color='b', label="v")
    plt.xlabel('Multi-channel Level')
    plt.ylabel('Intensity')
    plt.legend()
    plt.show()
    
```
