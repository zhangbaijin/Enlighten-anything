```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from random import random

# 创建画布
fig = plt.figure(figsize=(6, 10), facecolor='w')

# 创建 3D 坐标系
ax = fig.add_subplot(projection='3d')

# 定义数据
img = cv2.imread('/root/autodl-tmp/lighten-diffusion/23.png')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
x, y, z = img2[:, :, 0], img2[:, :, 1], img2[:, :, 2]

# 绘制 3D 散点
points = ax.scatter(xs=x,  # x 轴坐标
                    ys=y,  # y 轴坐标
                    zs=z,  # z 轴坐标
                    zdir='z',
                    s=1,  # size
                    alpha=0.05,
                    cmap='magma',
                    c=z)

mappable = plt.cm.ScalarMappable(cmap='magma')  # 色条
mappable.set_array(z)  # 色条以z为 orientation

# 设置坐标轴标题和刻度
c1 = ax.set(xlabel='H',
            ylabel='S',
            zlabel='V',  # 修改这里为你想要的标签文本
            xticks=np.arange(0, 255, 40),
            yticks=np.arange(0, 255, 40),
            zticks=np.arange(0, 255, 40)
            )

ax.set_xlabel('H', fontsize=15, labelpad=5)
ax.set_ylabel('S', fontsize=15, labelpad=5)
ax.set_zlabel('V', fontsize=11, labelpad=1)  # 修改这里为你想要的标签文本
labels = ax.get_xticklabels() + ax.get_yticklabels() + ax.get_zticklabels()

[label.set_fontsize(12) for label in labels]
[label.set_fontweight('bold') for label in labels]

# 调整视角
ax.view_init(elev=17, azim=45)

# 显示图形
plt.colorbar(mappable, fraction=0.02, cax=ax.inset_axes([1, 0.1, 0.05, 0.8]))  # 色条
plt.show()
```
